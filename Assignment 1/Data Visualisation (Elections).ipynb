```python
import matplotlib
import math
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_excel('/Users/dheem/Downloads/Jupyter/2014.xls',skiprows=3)
```

```python
data.columns
```

```python
data.describe()
```

```python
data_units = data[["STATE","POLL PERCENTAGE"]]

rep_plot = data_units.groupby('STATE').mean().plot(kind='bar')
rep_plot.set_xlabel("State")
rep_plot.set_ylabel("POLL PERCENTAGE")
```

```python
data_part = data[["STATE","Total voters"]]

rep_plot = data_part.groupby('STATE').mean().plot(kind='bar')
rep_plot.set_xlabel("State")
rep_plot.set_ylabel("Total voters")
```
